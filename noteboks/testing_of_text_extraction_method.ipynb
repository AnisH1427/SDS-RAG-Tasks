```python
import fitz
import pandas as pd
from tabulate import tabulate

pdf_document = fitz.open("/content/drive/MyDrive/RAG/2022 Q3 AAPL.pdf")

for page_num in range(pdf_document.page_count):
    page = pdf_document.load_page(page_num)
    tables = page.find_tables()

    if tables:
        print(f"Tables found on Page {page_num + 1}:")
        for i, table in enumerate(tables):
            print(f"Table {i+1} on Page {page_num + 1}:")
            try:
                # Extract data as a list of lists
                table_data = table.extract()
                # Convert to pandas DataFrame for better display
                df = pd.DataFrame(table_data)
                print(tabulate(df, headers='firstrow', tablefmt='grid'))
            except Exception as e:
                print(f"Could not extract data from table {i+1}: {e}")
            print("-" * 20)
    else:
        print(f"No tables found on Page {page_num + 1}")

pdf_document.close()
```

```python
import fitz
import pandas as pd

pdf_document = fitz.open("/content/drive/MyDrive/RAG/2022 Q3 AAPL.pdf")
extracted_data = {}

for page_num in range(pdf_document.page_count):
    page = pdf_document.load_page(page_num)

    # Extract all text from the page
    page_text = page.get_text("text")

    # Find tables on the page
    tables = page.find_tables()

    page_data = {
        "text": page_text,
        "tables": []
    }

    if tables:
        for i, table in enumerate(tables):
            try:
                # Extract table data
                table_data = table.extract()
                # Convert to pandas DataFrame
                df = pd.DataFrame(table_data)
                page_data["tables"].append({
                    # "table_number": i + 1,
                    # "bbox": table.bbox, # Store bounding box for potential context linking later
                    "data": df.to_dict('records') # Store as list of dictionaries for easy saving
                })
            except Exception as e:
                print(f"Could not extract data from table {i+1} on page {page_num + 1}: {e}")

    extracted_data[f"page_{page_num + 1}"] = page_data

pdf_document.close()

# Now, extracted_data dictionary contains text and table data for each page.
# You can inspect it or proceed to the next steps of the plan.
# For example, to see the data from page 4:
# print(extracted_data["page_4"])
```

```python
# Display the data extracted from page 4
display(extracted_data["page_6"])
```

```python
import json

# Define the path where you want to save the JSON file
output_json_path = "/content/drive/MyDrive/RAG/extracted_pdf_data.json"

# Save the extracted_data dictionary to a JSON file
with open(output_json_path, 'w') as f:
    json.dump(extracted_data, f, indent=4)

print(f"Extracted data saved to {output_json_path}")
```
